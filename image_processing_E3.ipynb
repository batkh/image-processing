```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read in the image
img = cv2.imread('/content/pollen-500x430px-96dpi.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization
equalized = cv2.equalizeHist(gray)

# Display the original and equalized images side-by-side
fig, axs = plt.subplots(1, 2)
axs[0].imshow(gray, cmap='gray')
axs[0].set_title('Original')
axs[1].imshow(equalized, cmap='gray')
axs[1].set_title('Equalized')
plt.show()
```

```python
# Define the scaling factor and shift factor
alpha =3  # increase the contrast
beta = 80  # increase the brightness

# Apply the linear transformation to the image
adjusted_img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

# Display the original and equalized images side-by-side
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img, cmap='gray')
axs[0].set_title('Original')
axs[1].imshow(adjusted_img, cmap='gray')
axs[1].set_title('adjusted_img')
plt.show()
```

```python
# Define the scaling factor and shift factor
alpha =1.05  # increase the contrast
beta =10  # increase the brightness
# Apply the linear transformation to the image
adjusted_img = cv2.convertScaleAbs(equalized, alpha=alpha, beta=beta)

# Display the original and equalized images side-by-side
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img, cmap='gray')
axs[0].set_title('Original')
axs[1].imshow(adjusted_img, cmap='gray')
axs[1].set_title('adjusted_img')
plt.show()
```

first of all we imply the histogram equalization in the orginal image after we convert it to gray scale to find the effection of the histogram equalization on the image after that we imply the 
cv2.convertScaleAbs() function. This function can be used to linearly adjust the pixel values of an image by scaling and shifting them.
then we apply the cv2.convertScaleAbs() function on the output of the histogram equalization function after we tweak convertScaleAbs() function paramters.the final result was much better we\hen we combine the effect of the two function.
