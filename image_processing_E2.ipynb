Propose a method for enhancing the image “nature_dark_forest.jpg” in terms of the
perceived light and color**

the method I propose is gamma_correction by using different values of gamma to get the desirable output 

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

original = cv2.imread('/content/nature_dark_forest.jpg')

def adjust_gamma(image, gamma=1.0):
    # build a lookup table mapping the pixel values [0, 255] to
    # their adjusted gamma values
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")

    # apply gamma correction using the lookup table
    return cv2.LUT(image, table)

fig, axes = plt.subplots(1, 3, figsize=(13,3))
plt.subplot(141), plt.imshow(original), plt.axis('off'), plt.title('Original')

gamma = 0.5
adjusted = adjust_gamma(original, gamma=gamma)
plt.subplot(142), plt.imshow(adjusted), plt.axis('off'), plt.title(f'Gamma = {gamma}')

gamma = 1.7
adjusted = adjust_gamma(original, gamma=gamma)
plt.subplot(143), plt.imshow(adjusted), plt.axis('off'), plt.title(f'Gamma = {gamma}')

gamma = 7
adjusted = adjust_gamma(original, gamma=gamma)
plt.subplot(144), plt.imshow(adjusted), plt.axis('off'), plt.title(f'Gamma = {gamma}')

plt.show()
```

after tryig many values of gamma the my observation found that the best value of gamm is between (1.7_7)
